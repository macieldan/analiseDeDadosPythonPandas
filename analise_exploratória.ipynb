```python
df=pd.read_excel('AdventureWorks.xlsx')
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df.shape
```

```python
#revceita total 
df['Valor Venda'].sum()
```

```python
df['custo'] = df["Custo Unit√°rio"].mul(df["Quantidade"])
```

```python
df['custo']
```

```python
#custo total 
round(df['custo'].sum(),2)
```

```python
#criando a coluna de lucro 
df['lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(1)
```

```python
df['lucro']
```

```python
#tempo necess√°rio para envio 
df['tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
df['tempo_envio'].dtype
```

```python
#media do tempo de envio por marca
df.groupby('Marca')['tempo_envio'].mean()
```

```python
#verificar valores ausentes 
df.isnull().sum()
```

```python
#agrupar por ano marca
df.groupby([df["Data Venda"].dt.year, 'Marca'])['lucro'].sum()
```

```python
#melhorando a vizualiza√ß√£o dos numeros 
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#agrupar por ano marca
df.groupby([df["Data Venda"].dt.year, 'Marca'])['lucro'].sum()
```

```python
#resetando o index 
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
```

```python
lucro_ano
```

```python
#total de produtos vendidos 
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Apresentar dados em gr√°fico
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title = "total produtos vendidos")
plt.xlabel('total')
plt.ylabel('produto');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title = 'lucro x ano')
plt.xlabel('ano')
plt.ylabel('receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
```

```python
#selecionando apenas os lucros de um ano espec√≠fico 
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['lucro'].sum().plot(title='lucro x mes')
plt.xlabel('mes')
plt.ylabel('lucro');
```

```python
df_2009.groupby('Marca')['lucro'].sum().plot.bar(title='lucro x marca')
plt.xlabel('marca')
plt.ylabel('lucro')
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby('Classe')['lucro'].sum().plot.bar(title='Lucro x Classe')
```
